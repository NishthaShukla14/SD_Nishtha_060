Expense Tracker


```python
class ExpenseTracker:
    # 1. Constructor: Yahan hum dictionary banayenge
    def __init__(self):
        # Dictionary jisme User ID aur Password store hain
        self.users_db = {
            "admin": "1234",
            "guest": "0000"
        }

    # 2. Login Method
    def login(self):
        print("--- Login System ---")
        user_id = input("Enter User ID: ")
        password = input("Enter Password: ")

        # Check karna ki ID dictionary mein hai ya nahi, aur password match kar raha hai ya nahi
        if user_id in self.users_db and self.users_db[user_id] == password:
            print("\nLogin Successful! Welcome.")
            self.calculate_expense() # Login sahi hone par expense wala method call hoga
        else:
            print("\nLogin Failed! Incorrect User ID or Password.")

    # 3. Expense Calculate karne ka Method
    def calculate_expense(self):
        print("\nPlease enter your expenses below:")
        food = int(input("Food Expense: ₹"))
        movie = int(input("Movie Expense: ₹"))
        travel = int(input("Travelling Expense: ₹"))
        other = int(input("Other Expense: ₹"))

        total = food + movie + travel + other

        # Summary Print Karna
        print("\n--- Your Expense Summary ---")
        print(f"🍔 Food: ₹{food}")
        print(f"🎬 Movie: ₹{movie}")
        print(f"🚌 Travelling: ₹{travel}")
        print(f"📦 Other: ₹{other}")
        print("----------------------------")
        print(f"💰 Total Expense: ₹{total}")

# 4. Class ka Object (Instance) banana aur program run karna
tracker = ExpenseTracker()
tracker.login()
```
