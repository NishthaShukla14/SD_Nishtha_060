```python
def count_test(n):
  count =1
  while count <= n:
    yield count
    count = count *3
```

```python
c= count_test(15)
```

```python
def count_digits(n):
  if n == 0:
    return 1
  else :
    return len(str(abs(n)))
```

```python
c =count_digits(1234)
```

```python
print(c)
```
