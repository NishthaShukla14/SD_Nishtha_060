```python
Student_name=input("Enter your name: ")
print(Student_name)

Sub_1=input("Enter your subject 1 name: ")
Mark_1=int(input(f"Enter marks for {Sub_1}: "))
print(f"{Sub_1}: {Mark_1}")
if Mark_1 >= 90:
    Grade = "A+"
elif Mark_1 >= 80:
    Grade = "A"
elif Mark_1 >= 70:
    Grade = "B"
elif Mark_1 >= 60:
    Grade = "C"
print(f"Congratulations! Your Grade is: {Grade}")

Sub_2=input("Enter your subject 2 name: ")
Mark_2=int(input(f"Enter marks for {Sub_2}: "))
print(f"{Sub_2}: {Mark_2}")
if Mark_2 >= 90:
    Grade = "A+"
elif Mark_2 >= 80:
    Grade = "A"
elif Mark_2 >= 70:
    Grade = "B"
elif Mark_2>= 60:
    Grade = "C"
print(f"Congratulations! Your Grade is: {Grade}")

Sub_3=input("Enter your subject 3 name: ")
Mark_3=int(input(f"Enter marks for {Sub_3}: "))
print(f"{Sub_3}: {Mark_3}")
if Mark_3>= 90:
    Grade = "A+"
elif Mark_3>= 80:
    Grade = "A"
elif Mark_3>= 70:
    Grade = "B"
elif Mark_3>= 60:
    Grade = "C"
print(f"Congratulations! Your Grade is: {Grade}")


Total_Marks = Mark_1 + Mark_2 + Mark_3
print(f"\nTotal Marks Obtained: {Total_Marks} out of 300")
Percentage = (Total_Marks / 300) * 100
print(f"Percentage: {Percentage}%")


CGPA = Percentage / 9.5
print(f"Your CGPA is:{round(CGPA, 2)}")
```
