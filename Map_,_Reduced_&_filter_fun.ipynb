```python
l = [2,3,4,5,6]
```

```python
def test(l):
  l1 = []
  for i in l:
    l1.append(i**2)
  return l1
```

```python
test(l)
```

```python
def sq(x):
  return x**2
```

```python
list(map(sq, l))
```

```python
list(map(lambda x : x**2 , l))
```

```python
l1 =[1,2,3,4,5]
l2 =[6,7,8,9,10]
list(map(lambda x , y : x+y , l1 , l2))
```

```python
def add(x,y) :
  return x+y
```

```python
list(map(add , l1 , l2))
```

```python
s = "skills"
list(map(lambda s : s.upper() , s))
```

```python
from functools import reduce
```

```python
l =[1,2,3,4,5]
reduce(lambda x , y : x+y , l)
```

```python
reduce(lambda x , y : x+y , l)
```

```python
reduce(lambda x , y : x*y , l)
```

```python
reduce(lambda x , y : x if x > y else y , l)
```

```python
# filter function
list(filter(lambda x : x % 2 == 0 , l))
```

```python
list(filter(lambda x : x % 2 != 0 , l))
```
